```python
# Importing Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers
import os
```

```python
# Define Paths

data_train_path = "D:/Programming/Textile ML Classification/Textile Classification/Train"
data_test_path = "D:/Programming/Textile ML Classification/Textile Classification/Test"
data_val_path = "D:/Programming/Textile ML Classification/Textile Classification/Validation"
```

```python
# Define Image Size 

img_height = 180
img_width = 180
```

```python
data_train = tf.keras.utils.image_dataset_from_directory(
    data_train_path,
    shuffle=True,
    image_size=(img_height, img_width),
    batch_size=32,
    validation_split=False)
```

```python
data_cat =data_train.class_names
```

```python
data_test = tf.keras.utils.image_dataset_from_directory(
    data_test_path,
    shuffle=False,
    image_size=(img_height, img_width),
    batch_size=32,
    validation_split=False)
```

```python
data_val = tf.keras.utils.image_dataset_from_directory(
    data_val_path,
    shuffle=False,
    image_size=(img_height, img_width),
    batch_size=32,
    validation_split=False)
```

```python
plt.figure(figsize=(10, 10))
for images, labels in data_train.take(1):
  for i in range(9):
    ax = plt.subplot(3, 3, i + 1)
    plt.imshow(images[i].numpy().astype("uint8"))
    plt.title(data_cat[labels[i]])
```
